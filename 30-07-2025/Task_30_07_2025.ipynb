```python
import pandas as pd
import matplotlib.pyplot as plt
# Step 1: Create CSV
csv_data = """Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000

Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
"""
with open('sales_data.csv', 'w') as file:
  file.write(csv_data)
# Step 2: Read the CSV
df = pd.read_csv('sales_data.csv')
print(df)
```

# Group the data by Store and calculate total sales for each store.

```python
total_sales=df.groupby('Store')['Sales'].sum()
print(total_sales)
```

# Identify which store had higher total sales.

```python
max_sales=total_sales.idxmax()
print("maximum Sales :",max_sales)

max=df[df['Sales']==df['Sales'].max()]
print("Maximum Sales per month :",max)
```

# 3. Monthly Sales Trend
Group the data by Month and sum up the total sales (Store A + Store B).
Print the result.

```python
total_month =df.groupby(['Month','Store'])['Sales'].sum()
print(total_month)
```

# 4. Chart: Monthly Sales by Store

```python
import matplotlib.pyplot as plt
pivot_df = df.pivot(index='Month', columns='Store', values='Sales')

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(pivot_df.index, pivot_df['Store A'], label='Store A', marker='o', linestyle='-', color='blue')
plt.plot(pivot_df.index, pivot_df['Store B'], label='Store B', marker='s', linestyle='--', color='green')

plt.title('Monthly Sales by Store')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df['Month'],df['Sales'],color="Teal")
plt.title("Monthy Sales")
plt.xlabel('Sales')
plt.ylabel('Month')
plt.grid(True)
plt.show()
```
