## The Head of the csv

```python
import pandas as pd

df=pd.read_csv('students.csv')
print(df.head(3))
```

# Average Of each student


```python
df.columns=df.columns.str.strip()
df['Average']=df[['Math','Science','English']].mean(axis=1)

print(df)
```

# Average of each subject


```python
df.columns=df.columns.str.strip()
avg=df[['Math','Science','English']].mean(axis=0)

print(avg)
```

# Max of Average


```python
avg_marks=df[df['Average']==df['Average'].max()]
print(avg_marks)
```

# Scored about 80 in Science


```python
high_scorer_science=df['Science']>80
print(df[high_scorer_science])
```

# Students with an average above 70


```python
avg=df['Average']>70
print(df[avg])
```

# Sorting The scores in descending order and Printing Top 3

```python
desc_English=df.sort_values('English',ascending=False)
print(desc_English.head(3))
```

# Bar Graph


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df['Name'],df['Average'],color='Black')
plt.title("Average of each student")
plt.xlabel("Name")
plt.ylabel("Average")
plt.grid(True)
plt.show()
```

# Line Graph


```python

df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='line', marker='o', figsize=(10, 6))

plt.title('Scores in Math, Science, and English')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.ylim(0, 100)
plt.grid(True)
plt.legend(title='Subjects')
plt.show()
```

# Highest Scorer among all students


```python
total_sum =df[['Math','Science','English']].sum()

highest_Subject=total_sum.idxmax()
highest_marks=total_sum.max()

print(f"The subject with highest mark {highest_Subject} ={highest_marks}")
```

```python

df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)


df.to_csv('students_summary.csv', index=False)

print("\nstudents_summary.csv has been saved successfully.")
```
